```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Path to the CSV file
file_path = "/Users/boltonhowes/Documents/NEOM/precipitation/"
fname1 = "eilat_man_precip.csv"
fname2 = "BPW_Time_Series_Data.csv"

# Read the CSV file into a DataFrame
eilat = pd.read_csv(file_path+fname1)
brine = pd.read_csv(file_path+fname2)
brine['Date'] = pd.to_datetime(brine['Date'])

# Display the first few rows of the DataFrame
print(eilat.head())
```

```python
# Convert the 'Date' column to datetime
eilat['Date'] = pd.to_datetime(eilat['Date'], format='%d/%m/%Y')

# Generate a complete date range
all_dates = pd.date_range(start=min(eilat['Date']), end=max(eilat['Date']))

# Create a new DataFrame with the complete date range
new_eilat = pd.DataFrame(all_dates, columns=['Date'])

# Merge the new DataFrame with the original DataFrame
new_eilat = pd.merge(new_eilat, eilat, on='Date', how='left')

# Fill missing values in 'Station', 'Daily Rainfall (mm)', 'Rain Code (code)' columns
new_eilat['Station'].fillna(method='ffill', inplace=True)
new_eilat['Daily Rainfall (mm)'].fillna(0, inplace=True)
new_eilat['Rain Code (code)'].fillna(0, inplace=True)

eilat = new_eilat
```

The eilat dataset starts before the brine pool, so lets truncate the eilat dataset

```python

# Convert Series to numpy array before plotting
dates = np.array(brine['Date'])
weighted_average = np.array(brine['Weighted Average'])

# Create a figure
plt.figure(figsize=(10, 10))

# Set the subplot for the Brine data
ax2 = plt.subplot(2, 1, 2)

# Plot the 'Weighted Average' column against 'Date'
ax2.plot(dates, weighted_average, color='black', linewidth=0.5)  # Use a thin line width

# Plot a blue dot where there is a flood in the Wadis
max_date = dates[np.argmax(weighted_average)]
ax2.scatter(max_date, np.max(weighted_average), color='blue', s=20)  # s is the size of the point

# Plot an empty blue circle at the 'Weighted Average' of '2019-01-27'
specific_date = '2019-01-27'
specific_weighted_avg = brine[brine['Date'] == specific_date]['Weighted Average'].values[0]
ax2.scatter(pd.to_datetime(specific_date), specific_weighted_avg, facecolors='none', edgecolors='blue', linewidth=0.5)


# Set the x-limit to show dates after 2000
ax2.set_xlim(pd.Timestamp('2000-01-01'), pd.Timestamp('2021-12-31'))

# Set y-limit
ax2.set_ylim([0, 45])

# Remove x and y ticks
ax2.set_yticks([10, 20, 30, 40])

# Remove the spines
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.spines['left'].set_visible(True)
ax2.spines['bottom'].set_visible(True)

# Set the axis ticks to be inside
ax2.tick_params(direction='in', length=4, width=0.5, colors='black')

# Add a faint grid
ax2.grid(True, linestyle='-', linewidth=0.25, color='gray', alpha=0.5)


plt.xlabel('Date', Fontsize = 14)
plt.ylabel('Daily Rainfall [mm]', Fontsize = 14)
# Add text to the plot
ax2.text(0.03, 0.89, 'Brine Pool', transform=ax2.transAxes, fontsize=16, color='black')


# Find the dates and values for the 10 highest rainfall
sorted_indices = np.argsort(weighted_average)[-10:][::-1]  # Get indices of the 10 highest rainfall
top_dates = dates[sorted_indices]
top_values = weighted_average[sorted_indices]

# Annotate the dates with the 10 highest rainfall
for i, (date, value) in enumerate(zip(top_dates, top_values)):
    date = pd.Timestamp(date)  # Convert numpy.datetime64 to pandas.Timestamp
    ax2.annotate(date.strftime('%Y-%m-%d'),  # Now you can use strftime
                 xy=(date, value),  # Position of the annotation
                 xytext=(5, 5),  # Offset text by 5 points to the right and up
                 textcoords='offset points',  # Coordinate system for text offset
                 fontsize=8, color='black',
                 rotation=45)  # Rotate the text by 45 degrees
    ax2.scatter(date, value, facecolor = 'none', edgecolor = 'black')  # Now you can use strftime



# Save the figure as a .svg file
plt.savefig('/Users/boltonhowes/Documents/NEOM/precipitation/brine_pool_rainfall.eps')

# Show the plot
plt.show()
```
