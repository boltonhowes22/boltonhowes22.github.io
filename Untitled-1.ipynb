```python
def adjust_longitude(gdf):
    """
    Adjusts the longitudes of a GeoDataFrame from [-180째, 180째] to [0째, 360째].
    """

    def adjust_coords(coords):
        return [(x + 360 if x < 0 else x, y) for x, y in coords]

    def adjust_polygon(polygon):
        outer_ring = adjust_coords(polygon.exterior.coords[:])
        inner_rings = [adjust_coords(ring.coords[:]) for ring in polygon.interiors]
        return type(polygon)(outer_ring, inner_rings)

    def adjust_geom(geom):
        if geom.type == 'Polygon':
            return adjust_polygon(geom)
        elif geom.type == 'MultiPolygon':
            return type(geom)([adjust_polygon(p) for p in geom])
        else:
            return geom

    gdf.geometry = gdf.geometry.apply(adjust_geom)
    return gdf

def create_100km_buffer_around_centroid(shapefile_path):
    """
    Create a bounding box around the centroid of the given shapefile.
    """
    gdf = gpd.read_file(shapefile_path).to_crs(epsg=4326)
    centroid = gdf.unary_union.centroid
    bounding_box = centroid.buffer(.25).envelope  # Buffer by 100 km (100,000 meters)
    return bounding_box

def mask_by_100km_box(geotiff_path, shapefile_path):
    """
    Mask a GeoTIFF by a box around the centroid of a given shapefile.
    """
    bounding_box = create_100km_buffer_around_centroid(shapefile_path)
    
    with rasterio.open(geotiff_path) as src:
        # Reproject the bounding box to match the GeoTIFF's CRS
        src_crs = src.crs
        bounding_box_transformed = gpd.GeoDataFrame({'geometry': [bounding_box]}, crs="EPSG:4326").to_crs(src_crs)
        bounding_box_transformed = bounding_box_transformed.geometry.iloc[0]
        
        cropped_img, cropped_transform = mask(src, [mapping(bounding_box_transformed)], crop=True)
    return cropped_img, cropped_transform, bounding_box


def rescale_band(band_data, old_min, old_max, new_min=0, new_max=255):
    return ((band_data - old_min) / (old_max - old_min) * (new_max - new_min) + new_min).astype(np.uint8)

def plot_geotiff_with_shapefile_boundary(ax, band_data, geotiff_transform, bounding_box, shapefile_path, geotiff_crs, vmin=None, vmax=None):
    # Read the shapefile and convert to the CRS of the geotiff
    gdf = gpd.read_file(shapefile_path)
    gdf = gdf.to_crs(geotiff_crs)  # Reproject using the CRS of the GeoTIFF

    # Overlay the shapefile boundary
    gdf.boundary.plot(ax=ax, color='red', linewidth=1)
    
    extent = rasterio.plot.plotting_extent(band_data, geotiff_transform)

    img_display = ax.imshow(band_data, cmap='gray', vmin=vmin, vmax=vmax, 
                            extent=extent, origin='lower')
    
    ax.set_xlim(extent[0], extent[1])
    ax.set_ylim(extent[3], extent[2])

    # Display lat/long gridlines
    ax.grid(True, which='both', linestyle='--', linewidth=0.5)
    ax.set_xlabel("Longitude",fontproperties=regular, fontsize=12)
    ax.set_ylabel("Latitude",fontproperties=regular, fontsize=12)
 
    
    return img_display


def extract_pixel_values(geotiff_path, shapefile_path, show=True):
    """
    Extract and normalize pixel values from a GeoTIFF within the bounds of a shapefile. 
    Optionally visualizes the GeoTIFF with the masked region.
    """
    geotiff_data, geotiff_transform, bounding_box = mask_by_100km_box(geotiff_path, shapefile_path)
    geotiff_data = np.clip(geotiff_data, 0, 4500)

    with fiona.open(shapefile_path, "r") as shapefile:
        shapes = [feature["geometry"] for feature in shapefile]

    pixel_values_list = []
    num_bands = geotiff_data.shape[0]

    with rasterio.open(geotiff_path) as src:
        geotiff_crs = src.crs

        # Convert Shapely geometries to a GeoDataFrame
        gdf = gpd.GeoDataFrame({'geometry': shapes}, crs="EPSG:4326") # Setting a default CRS

        # Transform the CRS of the GeoDataFrame
        gdf_transformed = gdf.to_crs(geotiff_crs)

        # Extract transformed Shapely geometries from the GeoDataFrame
        shapes_transformed = gdf_transformed['geometry'].tolist()

        for band in range(num_bands):
            # Extract pixel values for the given band
            band_data = geotiff_data[band]
            
            # Create a mask for the shapefile's geometry with correct dimensions
            geom_mask = rasterio.features.geometry_mask([mapping(shape) for shape in shapes_transformed], transform=geotiff_transform, invert=True, out_shape=band_data.shape)
            
            masked_band_data = band_data[geom_mask]
            pixel_values_list.append(masked_band_data)
    
    # Rest of the plotting code...
    
    # Get aspect ratio from the geotiff_data for the first band
    height, width = geotiff_data[0].shape
    aspect_ratio = height / width

    # Adjusting width for colorbar and spacing
    plots_width = 7.5
    fig_height = (plots_width / num_bands) * aspect_ratio

    if show:
        fig, axs = plt.subplots(1, num_bands, figsize=(15, fig_height))
        if num_bands == 1:
            axs = [axs]  # Ensure axs is always a list, even for one band

        img_displays = []  # Store all the img_display outputs

        # Calculate global minimum and maximum across all bands
        global_min = np.min(geotiff_data)
        global_max = np.max(geotiff_data)

        for band, ax in zip(range(num_bands), axs):
            band_data = geotiff_data[band]
            img_display = plot_geotiff_with_shapefile_boundary(ax, band_data, geotiff_transform,
                                                                bounding_box, shapefile_path, geotiff_crs, 
                                                                vmin=global_min, vmax=global_max)
            img_displays.append(img_display)

        # Find the first non-None img_display
        valid_img = next((img for img in img_displays if img is not None), None)
        if valid_img:
            cbar_ax = fig.add_axes([0.91, 0.18, 0.02, 0.7])
            fig.colorbar(valid_img, cax=cbar_ax)
            
        plt.tight_layout(rect=[0, 0, 0.9, 1])
        plt.show()

    return pixel_values_list

def display_3chan(image_path, channels=(0,1,2)):
    """
    Display an RGB image from a multi-channel GeoTIFF.

    Parameters:
    - image_path (str): Path to the GeoTIFF image.
    - channels (tuple or list): Indices of the channels to be displayed as RGB. Default is (0,1,2).

    Returns:
    - None. Displays the image using matplotlib.
    """
    with rasterio.open(image_path) as src:
        # Read the specified channels from the image
        r, g, b = (src.read(k+1) for k in channels)

        # Stack the channels along the third dimension
        rgb = np.stack((r, g, b), axis=-1)
        
        # Normalize for visualization
        rgb_norm = (rgb - rgb.min()) / (rgb.max() - rgb.min())

    # Display the RGB image using matplotlib
    plt.imshow(rgb_norm)
    plt.axis('off')  # Turn off axis numbers and ticks
    #plt.show()


def save_pixel_values_as_tiff(pixel_values, save_path):
    # Convert lists to numpy arrays
    if isinstance(pixel_values, list):
        pixel_values = np.array(pixel_values)

    # If it's a 1D array, let's reshape it to 2D with one row
    if len(pixel_values.shape) == 1:
        pixel_values = pixel_values.reshape(1, -1)

    with rasterio.open(save_path, 'w', driver='GTiff', height=pixel_values.shape[0],
                       width=pixel_values.shape[1], count=1, dtype=pixel_values.dtype,
                       crs='+proj=latlong') as dst:
        dst.write(pixel_values, 1)

def process_all_geotiffs_in_directory(directory_path, shapefile_path):
    """
    Extract and normalize pixel values from all GeoTIFFs within a directory using the bounds of a shapefile. 
    Optionally visualizes each GeoTIFF with the masked region.
    
    Args:
    - directory_path (str): Path to the directory containing GeoTIFF files.
    - shapefile_path (str): Path to the shapefile for masking the GeoTIFFs.

    Returns:
    - dict: A dictionary where keys are GeoTIFF filenames and values are the extracted pixel values.
    """
    pixel_values_dict = {}
    
    # Create the directory to save .tiff files if it doesn't exist
    save_directory = os.path.join(directory_path, 'region')
    if not os.path.exists(save_directory):
        os.makedirs(save_directory)

    for filename in os.listdir(directory_path):
        if filename.endswith(".tif") or filename.endswith(".tiff"):
            geotiff_path = os.path.join(directory_path, filename)
            
            # Extract pixel values using the provided function
            pixel_values = extract_pixel_values(geotiff_path, shapefile_path, show=False)
            
            pixel_values_dict[filename] = pixel_values
            
            # Save the pixel values as .tiff
            save_path = os.path.join(save_directory, filename)
            save_pixel_values_as_tiff(pixel_values, save_path)

    return pixel_values_dict


def convert_filename_format(filename):
    # Extract year, day number, and suffix from filename
    year = int(filename[:4])
    day_number = int(filename[4:7])
    suffix = filename[7:filename.find('.tif')]
    
    # Convert to date format
    date_obj = datetime(year, 1, 1) + timedelta(day_number - 1)
    
    new_filename = date_obj.strftime('%Y-%m-%d') + suffix + '.tif'
    return new_filename

def rename_files_in_directory(directory):
    # List all files in the directory
    files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
    
    # Loop over all files
    for filename in files:
        # Check if the file has the expected format
        if filename.endswith('_MYD09GQ.tif'):
            # Generate the new filename
            new_filename = convert_filename_format(filename)
            # Rename the file
            os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))
            print(f"Renamed {filename} to {new_filename}")

def generate_filename_array(directory):
    # List all files in the directory
    files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f)) and f.endswith('_MYD09GQ.tif')]
    
    return np.array([convert_filename_format(filename) for filename in files])
```

```python
def extract_date_from_modis_filename(filename):
    """Extracts the date from a MODIS filename."""
    try:
        year = int(filename[9:13])
        day_number = int(filename[13:16])
        # Convert to date format
        date_obj = datetime(year, 1, 1) + timedelta(day_number - 1)
        return date_obj
    except ValueError:
        print(f"Warning: Could not extract date from filename '{filename}'.")
        return None

def collect_files_in_directory(directory):
    """Collect all the files in the specified directory."""
    return [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]

geotiff_path = "/Users/boltonhowes/Documents/red_sea/code/modis_images/MYD09GA/geotiffs/"
all_modis_files = collect_files_in_directory(geotiff_path)
dates = [extract_date_from_modis_filename(f) for f in all_modis_files]
dates = [d for d in dates if d] 


# 'dates' is the list of dates extracted from the filenames
dates = sorted([extract_date_from_modis_filename(f) for f in all_modis_files if extract_date_from_modis_filename(f) is not None])

# Generate a continuous range of dates spanning the entire period
all_dates = pd.date_range(start=min(dates), end=max(dates))


fig, ax = plt.subplots(figsize=(10, 2))

for date in all_dates:
    if date in dates:
        ax.plot(date, 1, 'ko', markersize=5)  # black closed circle
    else:
        ax.plot(date, 0, 'ro', markerfacecolor='none', markersize=5)  # red open circle

ax.set_yticks([0, 1])
ax.set_xlabel('Date')
ax.set_title('Temporal Coverage of MODIS Data')
plt.tight_layout()
plt.show()
```
